```python
import pandas as pd  #  pandas 
import matplotlib.pyplot as plt  #  matplotlib 
import numpy as np
from google.colab import drive  #  google.colab  Google Colab 
drive.mount('/content/drive')  #  Colab  Google 

file_name = "/content/drive/MyDrive/^TWII.csv"  # 
df = pd.read_csv(file_name)  #  pandas  CSV  DataFrame 
window_size = 200  #  200

#  '200_Day_MA'NaN
df['200_Day_MA'] = np.nan

# 200200
for i in range(window_size - 1, len(df)):
    df.iloc[i, df.columns.get_loc('200_Day_MA')] = df['Close'][i - window_size + 1:i + 1].mean()





#  12x6
plt.figure(figsize=(12, 6))

#  'Date'  'Close'  "Close"
plt.plot(df['Date'], df['Close'], label="Close", color='blue')

#  'Date'  '200_Day_MA'  "200MA"
plt.plot(df['Date'], df['200_Day_MA'], label="200MA", color='red')

#  x  "Date"
plt.xlabel('Date')

#  y  "Price"
plt.ylabel('Price')

#  "Close_and_200MA"
plt.title('Close_and_200MA')

#  "Close"  "200MA" 
plt.legend()

# 
plt.grid(True)

# 
plt.show()
```
