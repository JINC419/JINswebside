```python
import pandas as pd  # 導入 pandas 函式庫，用於處理數據表格
import matplotlib.pyplot as plt  # 導入 matplotlib 函式庫，用於繪製圖表
from google.colab import drive  # 導入 google.colab 函式庫，用於連接到 Google Colab 的雲端
drive.mount('/content/drive')  # 將 Colab 連接到 Google 雲端硬碟

file_name = "/content/drive/MyDrive/^TWII.csv"  # 設定數據文件的路徑
df = pd.read_csv(file_name)  # 使用 pandas 讀取 CSV 文件，並將數據存儲到 DataFrame 中
window_size = 200  # 設定滑動窗口的大小為 200

# 計算並添加一列 '200_Day_MA' 到 DataFrame 中，該列包含了 'Close' 列的 200 日移動平均值
df['200_Day_MA'] = df['Close'].rolling(window=window_size).mean().dropna()

# 創建一個新的圖表，設定圖表的大小為 12x6
plt.figure(figsize=(12, 6))

# 繪製 'Date' 列與 'Close' 列的折線圖，設定線的標籤為 "Close"，線的顏色為藍色
plt.plot(df['Date'], df['Close'], label="Close", color='blue')

# 繪製 'Date' 列與 '200_Day_MA' 列的折線圖，設定線的標籤為 "200MA"，線的顏色為紅色
plt.plot(df['Date'], df['200_Day_MA'], label="200MA", color='red')

# 設定 x 軸標籤為 "Date"
plt.xlabel('Date')

# 設定 y 軸標籤為 "Price"
plt.ylabel('Price')

# 設定圖表的標題為 "Close_and_200MA"
plt.title('Close_and_200MA')

# 顯示圖例，它包含 "Close" 和 "200MA" 的標籤
plt.legend()

# 啟用圖表的網格線
plt.grid(True)

# 顯示圖表
plt.show()
```
